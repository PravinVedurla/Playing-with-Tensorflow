```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
house_data = pd.read_csv('cal_housing_clean.csv')
```

```python
house_data.columns
```

```python
X_val = house_data[['housingMedianAge', 'totalRooms', 'totalBedrooms', 'population',
       'households', 'medianIncome']]
```

```python
X_val.head()
```

```python
X_val.shape
```

```python
y_val = house_data['medianHouseValue']
```

```python
y_val.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_val, y_val, test_size=0.3 )
```

```python
X_train.shape
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(data=scaler.transform(X_train),
                      columns=X_train.columns,
                      index=X_train.index)
```

```python
X_train.head()
```

```python
X_test = pd.DataFrame(data=scaler.transform(X_test),
                      columns=X_test.columns,
                      index=X_test.index)
```

```python
X_test.head()
```

```python
house_data.columns
```

```python
age = tf.feature_column.numeric_column('housingMedianAge')
room = tf.feature_column.numeric_column('totalRooms')
bed = tf.feature_column.numeric_column('totalBedrooms')
pop = tf.feature_column.numeric_column('population')
hold = tf.feature_column.numeric_column('households')
medin =tf.feature_column.numeric_column('medianIncome')
```

```python
fea_cols= [age,room,bed,pop,hold,medin]
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=10,num_epochs=1000, shuffle=True)
```

```python
model = tf.estimator.DNNRegressor(hidden_units=[10,10,10],feature_columns=fea_cols)
```

```python
model.train(input_fn=input_func, steps=1000)
```

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test,batch_size=10, num_epochs=1, shuffle=False)
```

```python
list(model.predict(pred_input_func))
```
